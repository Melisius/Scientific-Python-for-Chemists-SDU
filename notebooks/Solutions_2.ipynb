### Ex 2.4

```python
def three_largest(dictionary):
    values = []
    for key in dictionary:
        values.append(dictionary[key])
    values.sort(reverse=True)
    return values[0:3]

dic = {"A": 10, "B": 20, "C": 300, "D": 40, "E": 61, "F": 60}
three_largest(dic)
```

### Ex 2.5


```python
def fibonacci(x, y):
    fib_numbers = [0, 1]
    while fib_numbers[-1] < y:
        if fib_numbers[-1] >= x:
            print(fib_numbers[-1])
        fib_numbers.append(fib_numbers[-1] + fib_numbers[-2])

fibonacci(10, 200)
```

### Ex 2.6

```python
def fizzbuzz(x):
    for i in range(1, x+1):
        if i % 3 == 0 and i % 5 == 0:
            print("fizzbuzz")
        elif i % 3 == 0:
            print("fizz")
        elif i % 5 == 0:
            print("buzz")
        else:
            print(i)
        
fizzbuzz(25)
```

### Ex 2.8

```python
def find_prime(x):
    i = 0
    prime_list = []
    while True:
        i += 1
        not_prime = 0
        for j in prime_list:
            if i%j == 0 and j != 1:
                not_prime = 1
                break
        if not_prime == 0:
            print(i)
            prime_list.append(i)
        
        if len(prime_list) == x:
            break

find_prime(10)
```

### Ex 2.9

```python
def factorial_recursion(n):
    if n < 1:
        return 1
    else:
        return n*factorial_recursion(n-1)
    
def factorial_loops(n):
    factorial = 1
    for i in range(1, n+1):
        factorial *= i
    return factorial
    
print(factorial_recursion(7))
print(factorial_loops(7))
```
