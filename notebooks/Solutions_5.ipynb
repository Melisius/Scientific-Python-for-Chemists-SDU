```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([1,2,3,4,5,6,7,8])
y1 = np.array([1,2,3,4,5,6,7,8])
x2 = x1*x1
y2 = y1

fig = plt.figure()
plot1 = fig.add_subplot(1,2,1)
plot1.plot(x1, y1, marker='x', linestyle='None', color='green')
plot1.set_title("Plot one")

plot2 = fig.add_subplot(1,2,2)
plot2.plot(x2, y2, marker='o', linestyle='--', color='magenta')
plot2.set_title("Plot two")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([1,2,3,4,5,6,7,8])
y1 = np.array([1,2,3,4,5,6,7,8])
x2 = x1*x1
y2 = y1

fig = plt.figure()
plot1 = fig.add_subplot(1,2,1)
plot1.plot(x1, y1, marker='x', linestyle='None', color='green')
plot1.set_title("Plot one")
plot1.set_xlabel("x axis of plot one")
plot1.set_ylabel("y axis of plot one")

plot2 = fig.add_subplot(1,2,2)
plot2.plot(x2, y2, marker='o', linestyle='--', color='magenta')
plot2.set_title("Plot two")
plot2.set_xlabel("x axis of plot two")
plot2.set_ylabel("y axis of plot two")

plt.show()
```

## Ex 5.4

```python
import matplotlib.pyplot as plt
import numpy as np

x11 = np.array([1,2,3,4,5,6,7,8])
y11 = np.array([1,2,3,4,5,6,7,8])
x12 = x11*x11
y12 = y11
x21 = x11/(1+x12)
y21 = y11
x22 = y11
y22 = x11*5

fig = plt.figure()
plot1 = fig.add_subplot(1,2,1)
plot1.plot(x11, y11, label="First curve")
plot1.plot(x12, y12, label="Second curve")
plot1.legend()

plot2 = fig.add_subplot(1,2,2)
plot2.plot(x21, y21, label="Third curve")
plot2.plot(x22, y22, label="Fourth curve")
plot2.legend()

plt.show()
```
